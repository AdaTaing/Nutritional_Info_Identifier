Adding folders to raw_images (if not existed)

```python
import os
```

```python
print(os.getcwd())
```

```python
# Create a folder named according to the fruits dictionary in the images_raw folder of the project
for fruit in fruits:
    #check if folder exists
    if not os.path.exists(f'../images_raw/{fruit}'):
        os.makedirs(f'../images_raw/{fruit}')
        print(f'Folder {fruit} created')
    else:
        print(f'Folder {fruit} already exists')
```

modify image names

```python
# modify the image names in the folder to have a prefix according to the name
food_name = 'fig'
for i, filename in enumerate(os.listdir(f'../images_raw/{food_name}')):
    src = f'../images_raw/{food_name}/{filename}'
    dst = f'../images_raw/{food_name}/{food_name}_{i+1}.jpg'
    os.rename(src, dst)
```

```python
# Create a folder named according to the vegetables dictionary in the images_raw folder of the project
for vege in vegetables:
    #check if folder exists
    if not os.path.exists(f'../images_raw/{vege}'):
        os.makedirs(f'../images_raw/{vege}')
    else:
        print(f'Folder {vege} already exists')
```
