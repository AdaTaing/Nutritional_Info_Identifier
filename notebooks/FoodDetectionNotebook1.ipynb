Adding folders to raw_images (if not existed)

```python
import os
```

```python
# Create a folder named according to the fruits dictionary in the images_raw folder of the project
for fruit in fruits:
    #check if folder exists
    if not os.path.exists(f'../images_raw/{fruit}'):
        os.makedirs(f'../images_raw/{fruit}')
    else:
        print(f'Folder {fruit} already exists')
```

modify apple image names

```python
# modify the image names in the 'apple' folder to have a prefix 'apple_'
for i, filename in enumerate(os.listdir('../images_raw/apple')):
    src = f'../images_raw/apple/{filename}'
    dst = f'../images_raw/apple/apple_{i+1}.jpg'
    os.rename(src, dst)
```

add empty/filler txt files to each fruit folder in the images folder

```python
# add an txt file in each fruit folder in the images folder with the name 'info.txt' and have the name of the fruit inside
for fruit in fruits:
    with open(f'../images_raw/{fruit}/info.txt', 'w') as f:
        f.write(fruit)
```
