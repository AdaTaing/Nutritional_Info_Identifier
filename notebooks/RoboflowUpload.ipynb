```python
# get list of images in apple folder
import os
food_folder = "cantaloupe"
image_files = os.listdir("../images_raw/" + food_folder + "/")
print(image_files)

#reorder the list to be in numerical order
image_files.sort(key=lambda x: int(x.split('_')[1].split('.')[0]))
print(image_files)
print(len(image_files))

#cut the list in half to only upload first half of images
image_files = image_files[:len(image_files)//2]
print(image_files)
print(len(image_files))
```

```python
from roboflow import Roboflow

API_KEY = "waJKh53lwJ6PGdpSRmGp"

# Initialize the Roboflow object with your API key
rf = Roboflow(api_key=API_KEY)

# Retrieve your current workspace and project name
print(rf.workspace())

# Specify the project for upload
# let's you have a project at https://app.roboflow.com/my-workspace/my-project
workspaceId = 'nutrition-identifier'
projectId = 'nutrition-identifier-x4sjn'
project = rf.workspace(workspaceId).project(projectId)

# Upload the image to your project
image_path = f"../images_raw/apple/apple_1141.jpg"
count = 0

#upload all images from the apple folder to the project
for image_file in image_files:
        image_path = f"../images_raw/{food_folder}/{image_file}"
        project.upload(image_path)
        count += 1

print(f"Uploaded {count} {food_folder} images to Roboflow project '{projectId}' in workspace '{workspaceId}'.")
```
